2. Отберем релевантные записи:

```python
columns_to_keep = [
    "Molecule ChEMBL ID", "Molecule Name", "Molecular Weight", "Smiles", "AlogP", 'Ligand Efficiency BEI', 'Ligand Efficiency LE', 'Ligand Efficiency LLE', 'Ligand Efficiency SEI',
    "Standard Type", "Standard Relation", "Standard Value", "Standard Units",
    "pChEMBL Value", "Target ChEMBL ID", "Target Name", "Target Organism", "Action Type", "AlogP", "Molecular Weight", "Ligand Efficiency LE"
]
```

```python
df_activity['Standard Value'] = pd.to_numeric(df_activity['Standard Value'], errors='coerce')
```

Теперь значения в столбце Standard Value - численные.

3. Приведем значения активности к единому формату

```python
def normalize_to_nM(value, unit):
    if pd.isna(unit):
        return None
    unit = str(unit).lower()
    if unit == "nm":
        return value
    elif unit in ["μm", "um"]:
        return value * 1000
    elif unit == "mm":
        return value * 1_000_000
    else:
        return None

df_activity["Standard Value"] = df_activity.apply(
    lambda row: normalize_to_nM(row["Standard Value"], row["Standard Units"]),
    axis=1
)

df_activity = df_activity[df_activity["Standard Value"].notna()]

```

```python
df_activity["Standard Value"]
```

4. Обработаем пропущенные значения и дубликаты:

```python
df_activity.isna().sum()
```

```python
df_activity = df_activity.dropna(subset=["Smiles", "Standard Value", "Standard Units"])
```

```python
df_activity.duplicated()
```

```python
duplicates = df_activity[df_activity.duplicated(keep=False)]
duplicates
```

```python
df_activity[df_activity.duplicated(subset=["Smiles"], keep=False)]
```

Удалим подные дупликаты

```python
df_activity = df_activity.drop_duplicates()
```

Для остальных дупликатов возьмем те, что с лучшей активностью:

```python
idx = df_activity.groupby('Molecule ChEMBL ID')['Standard Value'].idxmin()
df_activity = df_activity.loc[idx].reset_index(drop=True)
```

```python
df_activity
```

5. Проверим корректность молекул:

```python
def is_correct_smiles(smiles):
    try:
        mol = Chem.MolFromSmiles(smiles)
        return mol is not None
    except:
        return False
```

```python
df_activity["correct_smiles"] = df_activity["Smiles"].apply(is_correct_smiles)
```

```python
df_activity["correct_smiles"].sum()
```

Получается, что корректных смайлсов 5100.

```python
df_activity = df_activity[df_activity["correct_smiles"]].drop(columns=["correct_smiles"])
```

```python
invalid_count = (~df_activity["Smiles"].apply(is_correct_smiles)).sum()
print(f"Количество НЕвалидных SMILES: {invalid_count}")
```

```python
df_activity.to_csv("clean_activity.csv", index=False)
```
