```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
df = pd.read_csv('Bioactivity_data_Cyclooxygenase-2.csv', delimiter = ';')
df
```

```python
df.info()
```

```python
#Удаление столбцов не обладающих важной информацией об активности или имеющих большое количество пропусков
df = df.drop(df.columns[20:44], axis=1)
df = df.drop(['Standard Relation', 'Uo Units', 'Compound Key', 'Standard Type', 'Standard Text Value', 'Properties', 'Action Type', 'Data Validity Comment', 'Comment', 'Molecule Name', 'Molecule Max Phase'], axis=1)
df.info()
```

```python
# Удаление строк с пропусками. Возможно три подхода. 
# Первый, убрать Ligand Efficiency BEI/LE/LLE/SEI и тогда останется больше строк (информации), 
# но в тоже время потерятся информация по лигандной эффективности.
# Второй, убрать все строки с пропусками, так как дополнение средним, медианой не будет иметь никакого физического и химического смысла 
# и может привести к неточным предсказаниям моделей в дальнейшем. 
# Третий подход подразумевает расчёт Ligand Efficiency BEI/LE/LLE/SEI и дополнение датасета, 
# так же расчёт можно осуществить и для других дескрипторов (такой подход можно реализовать в дальнейшем при обучении моделей)
# Остановимся на втором варианте
clean_df = df.dropna(axis=0, how='any')
```

```python
#Проверка на различные размерности показала, что остались только соразмерные значения (nM)
clean_df['Standard Units'].value_counts()
```

```python
clean_df = clean_df.drop('Standard Units', axis=1)
clean_df
```

```python
#Было замечено, что для некоторых Molecule ChEMBL ID существует несколько значений
clean_df['Molecule ChEMBL ID'].value_counts()
```

```python
# Анализ распределения значнений для молекулы CHEMBL118. 
# В дальнейшем была осуществлена фильтрация по принципу выбора самого распространённого значения, в данном случае 40.0
CHEMBL118 = df[df['Molecule ChEMBL ID'] == 'CHEMBL118']
CHEMBL118['Standard Value'].value_counts()
```

```python
# Фильтрация для молекул имеющих несколько различных значений. Выбирается одно самое распространнённое, а его дубликаты и другие версии удаляются
idx = clean_df.groupby('Molecule ChEMBL ID')['Standard Value'].transform(
    lambda x: x == x.value_counts().idxmax())
pure_df = clean_df[idx].drop_duplicates(subset=['Molecule ChEMBL ID'])
```

```python
# Проверка
pure_df['Molecule ChEMBL ID'].value_counts()
```

```python
pure_df
```

```python
# Проверка на невалидные smiles путём создания mol файла, индексы некорректных smiles сохраняются для дальнейшего удаления
def valid_smiles(df, smiles_column='Smiles'):
    invalid_indices = []
    
    for idx, smiles in df[smiles_column].items():
        try:
            mol = Chem.MolFromSmiles(str(smiles))
            if mol is None:
                invalid_indices.append(idx)
        except:
            invalid_indices.append(idx)
    
    return invalid_indices

invalid_indices = valid_smiles(pure_df, 'Smiles')
# crystal_pure_df = pure_df.drop(index=invalid_indices) 
```
